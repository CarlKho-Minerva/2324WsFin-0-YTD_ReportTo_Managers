```python
# manager's email in one column and the formatted message in the other
import csv
import pandas as pd

# Read the CSV file into a DataFrame
sheet = pd.read_csv(
    "YTD and Spring semester total hours  - Total Hours Summa.csv")  # Replace with the path to your CSV file

# Create an empty dictionary to store messages for each manager
messages_dict = {}

# Group the DataFrame by "Manager Email"
grouped_sheet = sheet.groupby("Manager Email")

# Loop through each group
for email, group in grouped_sheet:
    # Get the manager's name and email address
    manager_name = group["Manager Name"].iloc[0]
    manager_email = email

    # Print a message to confirm the manager's name and email address
    print(f"Creating message for {manager_name} at {manager_email}...")

    # Loop through each intern in the group
    message = f"Dear {manager_name},\n\nWe hope you are doing well.\n\n\nBelow, you can find the Year-to-Date and Semester total hours of your WS interns.\n\n"

    for index, row in group.iterrows():
        student_first_name = row["Student First Name"]
        student_last_name = row["Student Last Name"]
        ytd = row["Year-To-Date Total Hours"]
        spring = row["Spring Total Hours"]
        left_hours = row["Total Spring hours left"]
        message += f"{student_first_name} {student_last_name}: YTD = {ytd}, Spring hours = {spring}, Hours left to work = {left_hours}\n\n"

    message += "Best,\nThe Student Finance Team"

    # Add the message to the messages_dict, with the manager's email as the key
    messages_dict[manager_email] = message

# Write the messages to a CSV file
with open("YTD_Report_SY2324.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Manager Email", "Message"])
    for email, message in messages_dict.items():
        writer.writerow([email, message])

print("CSV file created successfully.")
```

```python
#V2 - Run every saturday at 8AM
import csv
import pandas as pd
import schedule
import time

def generate_messages():
    # Read the CSV file into a DataFrame
    sheet = pd.read_csv("YTD and Spring semester total hours  - Total Hours Summa.csv")  # Replace with the path to your CSV file

    # Create an empty dictionary to store messages for each manager
    messages_dict = {}

    # Group the DataFrame by "Manager Email"
    grouped_sheet = sheet.groupby("Manager Email")

    # Loop through each group
    for email, group in grouped_sheet:
        # Get the manager's name and email address
        manager_name = group["Manager Name"].iloc[0]
        manager_email = email

        # Print a message to confirm the manager's name and email address
        print(f"Creating message for {manager_name} at {manager_email}...")

        # Loop through each intern in the group
        message = f"Dear {manager_name},\nI hope you are doing well.\n\nThere was a technical issue in the previous email. You can find the updated data in this email.\n\nBelow, you can find the Year-to-Date and Semester total hours of your WS interns.\n\n"
        for index, row in group.iterrows():
            student_first_name = row["Student First Name"]
            student_last_name = row["Student Last Name"]
            ytd = row["Year-To-Date Total Hours"]
            spring = row["Spring Total Hours"]
            left_hours = row["Total Spring hours left"]
            message += f"{student_first_name} {student_last_name}: YTD = {ytd}, Spring hours = {spring}, Hours left to work = {left_hours}\n\n"

        message += "Best,\nEmil Rustamli"

        # Add the message to the messages_dict, with the manager's email as the key
        messages_dict[manager_email] = message

    # Write the messages to a CSV file
    with open("messages.csv", "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["Manager Email", "Message"])
        for email, message in messages_dict.items():
            writer.writerow([email, message])

    print("CSV file created successfully.")

# Schedule the job to run every Saturday at 8:00 AM
schedule.every().saturday.at("08:00").do(generate_messages)

# Run the scheduler indefinitely
while True:
    schedule.run_pending()
    time.sleep(1)
```

```python
import csv
import pandas as pd
from datetime import datetime
from google.oauth2 import service_account
import gspread
import schedule
import time
from googleapiclient.discovery import build
from googleapiclient.http import MediaFileUpload

# Define Google Drive API credentials file path
credentials_file_path = "ytd-report-sy2324-5986472ba88e.json"  # Replace with your credentials file path

# Define Google Drive folder ID
folder_id = "194WGRJ7RcUChi-vhSCQsl7TD3SuTdrgz"  # Replace with your folder ID -- currently found in https://drive.google.com/drive/u/0/folders/194WGRJ7RcUChi-vhSCQsl7TD3SuTdrgz

# Read the CSV file into a DataFrame
sheet = pd.read_csv("YTD_Source_WS2223.csv")  # Replace with the path to your CSV file

# Create an empty dictionary to store messages for each manager
messages_dict = {}

# Group the DataFrame by "Manager Email"
grouped_sheet = sheet.groupby("Manager Email")

# Loop through each group
for email, group in grouped_sheet:
    # Get the manager's name and email address
    manager_name = group["Manager Name"].iloc[0]
    manager_email = email

    # Print a message to confirm the manager's name and email address
    print(f"Creating message for {manager_name} at {manager_email}...")

    # Loop through each intern in the group
    message = f"Dear {manager_name},\n\nWe hope you are doing well.\n\nBelow, you can find the Year-to-Date and Semester total hours of your WS interns.\n\n"

    for _, row in group.iterrows():
        student_first_name = row["Student First Name"]
        student_last_name = row["Student Last Name"]
        ytd = row["Year-To-Date Total Hours"]
        spring = row["Spring Total Hours"]
        left_hours = row["Total Spring hours left"]
        message += f"{student_first_name} {student_last_name}: YTD = {ytd}, Spring hours = {spring}, Hours left to work = {left_hours}\n\n"

    message += "Best,\nThe Student Finance Team"

    # Add the message to the messages_dict, with the manager's email as the key
    messages_dict[manager_email] = message

# Generate the filename with the current date as a prefix
current_date = datetime.now().strftime("%Y-%m-%d")
filename = f"{current_date}_YTD_ManagersReport_SY2324.csv"

# Write the messages to the CSV file
with open(filename, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Manager Email", "Message"])
    for email, message in messages_dict.items():
        writer.writerow([email, message])

print(f"CSV file '{filename}' created successfully.")

# Authenticate and create a connection to Google Drive API
credentials = service_account.Credentials.from_service_account_file(credentials_file_path, scopes=["https://www.googleapis.com/auth/drive"])
drive_service = build('drive', 'v3', credentials=credentials)

# Upload the CSV file to Google Drive
file_metadata = {
    "name": filename,
    "parents": [folder_id]
}

media = MediaFileUpload(filename, mimetype="text/csv")

try:
    sheet_file = drive_service.files().create(body=file_metadata, media_body=media).execute()
except Exception as e:
    print(f"An error occurred while uploading the file: {e}")


# Convert the CSV file to Google Sheets format
sheet_service = gspread.authorize(credentials)
spreadsheet = sheet_service.open_by_key(sheet_file["id"])
sheet = spreadsheet.sheet1
sheet.export(format="csv")

print("CSV file uploaded to Google Drive and converted to Google Sheets format.")
```
